#### Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
stats = pd.read_csv('ratings.csv')
def classifier(rating):
    
    if rating <= 2:
        return 'низкий рейтинг'
    elif 2 < rating <= 4:
        return 'средний рейтинг'
    elif rating == 4.5:
        return 'высокий рейтинг'
    elif rating == 5:
        return 'высокий рейтинг'
        
stats['class'] = stats['rating'].apply(classifier)
stats.tail(300)
```

```python
stat = pd.read_csv('ratings.csv')
def classif(stat_):
    
    if stat_['rating'] <= 2:
        return 'низкий рейтинг'
    elif 2 < stat_['rating'] <= 4:
        return 'средний рейтинг'
    elif stat_['rating'] == 4.5:
        return 'высокий рейтинг'
    elif stat_['rating'] == 5:
        return 'высокий рейтинг'
#     else:
#         return row
        
stat['class'] = stat.apply(classif, axis=1)
stat.tail(300)
```

```python
def geo_inf(inf_):
    for k, v in geo_data.items():
        for x in v:
            if x in inf_:
                return k
    else:
        return 'undefined'
```
